```python
import pandas as pd 
import numpy as np 
import time
from dt_ml.data_loader import load_csv 
from dt_ml.baseline_analytics import compute
print("All libraries loaded successfully")
```

```python
DATASETS = { "sales": "C:/Users/kdill/OneDrive/Attachments/dt_ml/validation/test_data/sales_data.csv", 
            "operations": "C:/Users/kdill/OneDrive/Attachments/dt_ml/validation/test_data/ops_data.csv", 
            "marketing": "C:/Users/kdill/OneDrive/Attachments/dt_ml/validation/test_data/marketing_data.csv",
            "customers": "C:/Users/kdill/OneDrive/Attachments/dt_ml/validation/test_data/customer_data.csv",
            "financials": "C:/Users/kdill/OneDrive/Attachments/dt_ml/validation/test_data/financials.csv",
            "hr": "C:/Users/kdill/OneDrive/Attachments/dt_ml/validation/test_data/hr_data.csv"
            }
```

```python
loaded_data = {}
print("=" * 70) 
print("LOADING DATASETS") 
print("=" * 70) 
for name, path in DATASETS.items(): 
    try:
         df = load_csv(path)
         loaded_data[name] = df 
         print(f"\n{name.upper()} DATASET LOADED SUCCESSFULLY") 
         print(f"Rows: {len(df)}") 
         print(f"Columns: {len(df.columns)}") 
         print("\nColumns:")
         print(df.columns.tolist()) 
         print("\nPreview:")
         print(df.head())
    except Exception as e: 
        print(f"\nFAILED TO LOAD {name}") 
        print(e)
```

```python
results = {} 
print("\n" + "=" * 70) 
print("RUNNING BASELINE ANALYTICS")
print("=" * 70) 
for name, df in loaded_data.items():
     try:
         print("\n" + "=" * 50) 
         print(f"DATASET: {name.upper()}") 
         print("=" * 50)
         result = compute(df)
         results[name] = result # KPI CARDS 
         print("\nKPI CARDS") 
         print("-" * 30) 
         for k, v in result["kpi_cards"].items():
             print(f"{k}: {v}") # GROWTH RATE 
             print("\nGROWTH RATE")
             print("-" * 30)
             print(result["growth_rate_pct"]) # TREND 
             print("\nTREND") 
             print("-" * 30) 
             print(result["trend"]) # CHURN RATE 
             print("\nCHURN RATE")
             print("-" * 30) 
             print(result["churn_rate_pct"])
             # CAC 
             print("\nAVG MARKETING CAC") 
             print("-" * 30) 
             print(result["avg_marketing_cac"]) 
             # HEADCOUNT COST 
             print("\nHEADCOUNT COST TOTAL") 
             print("-" * 30) 
             print(result["headcount_cost_total"]) 
             # QUALITY SCORE
             print("\nDATA QUALITY SCORE") 
             print("-" * 30) 
             print(result["data_quality_score"]) 
             # WARNINGS 
             print("\nWARNINGS") 
             print("-" * 30) 
             if len(result["warnings"]) == 0: 
                print("No warnings")
             else: 
                 for w in result["warnings"]:
                    print(f"- {w}") 
            # MONTHLY REVENUE
             print("\nMONTHLY REVENUE SAMPLE") 
             print("-" * 30) 
             print(result["monthly_revenue"][:5]) 
             # SEASONALITY print("\nSEASONALITY INDEX") 
             print("-" * 30) 
             print(result["seasonality_index"])
     except Exception as e:
         print(f"\nFAILED ANALYTICS FOR {name}") 
         print(e)
```

```python
print("\n" + "=" * 70) 
print("PERFORMANCE TEST")
print("=" * 70) 
try: 
    sales_df = loaded_data["sales"] 
    # Create large dataframe (~10K rows)
    big_df = pd.concat( [sales_df] * 100, ignore_index=True ) 
    print(f"Rows in test dataframe: {len(big_df)}")
    start = time.time() 
    compute(big_df) 
    end = time.time() 
    runtime = round(end - start, 2) 
    print(f"\nRuntime: {runtime} seconds") 
    if runtime < 2: 
        print("PASSED PERFORMANCE REQUIREMENT (<2 sec)") 
    else: 
        print("FAILED PERFORMANCE REQUIREMENT")
except Exception as e: 
    print("Performance test failed") 
    print(e)
```
